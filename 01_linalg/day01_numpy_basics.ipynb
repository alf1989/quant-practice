```python
import numpy as np

# 1. Create vectors
x = np.array([1.0, 2.0, 3.0])
w = np.array([0.2, 0.3, 0.5])

# 2. Weighted sum
weighted_sum = x @ w
print("Weighted sum:", weighted_sum)

# 3. Mean and standard deviation
x_mean = x.mean()
x_std = x.std()

# 4. Standardize
z = (x - x_mean) / x_std
print("Standardized x:", z)

# 5. Matrix example
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print("A @ B =")
print(A @ B)

# 6. Broadcasting example
v = np.array([10, 20])
print("A + v =")
print(A + v)

print("Shapes:")
print("x:", x.shape)
print("A:", A.shape)

print("Elementwise A * B =")
print(A * B)

print("Sample std of x:")
print(x.std(ddof=1))
```
