```python
import numpy as np

# -----------------------------
# 1. Vector dot product
# -----------------------------
u = np.array([1.0, 2.0, 3.0])
v = np.array([4.0, 5.0, 6.0])

dot_uv = u @ v
print("u @ v =", dot_uv)

# -----------------------------
# 2. Matrix-vector multiplication
# -----------------------------
M = np.array([[1.0, 2.0],
              [3.0, 4.0]])
x = np.array([1.0, 2.0])

Mx = M @ x
print("\nM @ x =")
print(Mx)

# -----------------------------
# 3. Matrix-matrix multiplication
# -----------------------------
N = np.array([[2.0, 0.0],
              [1.0, 2.0]])

MN = M @ N
print("\nM @ N =")
print(MN)

# -----------------------------
# 4. Transpose
# -----------------------------
print("\nM.T =")
print(M.T)

# -----------------------------
# 5. Solve Ax = b
# -----------------------------
A = np.array([[3.0, 1.0],
              [1.0, 2.0]])
b = np.array([9.0, 8.0])

solution = np.linalg.solve(A, b)
print("\nSolution to A x = b:")
print(solution)

print("\nCheck A @ solution =")
print(A @ solution)

# -----------------------------
# 6. Compare with inverse
# -----------------------------
A_inv = np.linalg.inv(A)
solution_via_inv = A_inv @ b

print("\nA inverse =")
print(A_inv)

print("\nSolution via inverse:")
print(solution_via_inv)

# -----------------------------
# 7. Shapes
# -----------------------------
print("\nShapes:")
print("u:", u.shape)
print("M:", M.shape)
print("x:", x.shape)
print("A:", A.shape)
print("b:", b.shape)
```
